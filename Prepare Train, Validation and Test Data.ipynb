```python
# create a test data folders which
# have both from and to tractors
# but do not have any of the images selected for train and validate

HAND_PICKED = r'C:\DS Project\_processed\Images\resized_255_\hand_picked'

# MIKE 
ALL_TRACTORS = r'C:\DS Project\_processed\Images\resized_255_\tractors\Mike'
DATA_ROOT = r'C:\DS Project\_for_keras\mike_from_to\data'    
HAND_PICKED_TO = HAND_PICKED + r'\to\mike'
HAND_PICKED_FROM = HAND_PICKED + r'\from\mike'

# NEIL
#ALL_TRACTORS = r'C:\DS Project\_processed\Images\resized_255_\tractors\Mike'
#DATA_ROOT = r'C:\DS Project\_for_keras\mike_from_to\data'    

#folder structure
TEST_DATA = DATA_ROOT + r'\test\combined'
TRAIN_DATA_CLASS_TO = DATA_ROOT + r'\train\to'
TRAIN_DATA_CLASS_FROM = DATA_ROOT + r'\train\from'
VALID_DATA_CLASS_TO = DATA_ROOT + r'\valid\to'
VALID_DATA_CLASS_FROM = DATA_ROOT + r'\valid\from'
```

```python
import os
import shutil 
import numpy as np
```

```python
def createFolder(folder):
    
    try:
        os.makedirs(folder)
    except OSError:
        print ("Creation of the directory %s failed" % folder)
    else:
        print ("Successfully created the directory %s " % folder)
   
          
def createTestData(ALL_TRACTORS, HAND_PICKED, TEST_DATA):
    
    #loop round all images and if not in 'hand picked' data then copy to test data
    for filename in os.listdir(ALL_TRACTORS):
    
        hand_picked = False
    
        #loop for all images manually selected for train/valid
        for subdir, dirs, files in os.walk(HAND_PICKED):
            for tractorFilename in files:
                if tractorFilename == filename:
                    print("dont copy",filename )
                    hand_picked = True
    
        if hand_picked == False:
            #copy file to test data
            src = ALL_TRACTORS + r'\\' + filename
            dest = TEST_DATA + r'\\' + filename
            #print (src, dest)
            shutil.copy(src, dest)
            
def createTrainAndValidateData(TRAIN_DATA_CLASS_TO, TRAIN_DATA_CLASS_FROM, VALID_DATA_CLASS_TO, VALID_DATA_CLASS_FROM):
    
    #split the To Digester class images into train and validate
    
    ###################
    # Class : To
    ###################
    allFileNames = os.listdir(HAND_PICKED_TO)
    np.random.shuffle(allFileNames)
    trainFileNames, validFileNames = np.split(np.array(allFileNames),[int(len(allFileNames)*0.5)])
  
    print('Total hand picked to: ', len(allFileNames))
    print('Training: ', len(trainFileNames))
    print('Validation: ', len(validFileNames))

    # Copy-images to correct folder (train, valid) for this class
    # TRAIN TO
    for name in trainFileNames:
        src = HAND_PICKED_TO + r'\\' + name
        dest = TRAIN_DATA_CLASS_TO + r'\\' + name
        print(src,dest)
        shutil.copy(src,dest)

    # VALID TO    
    for name in validFileNames:
        src = HAND_PICKED_TO + r'\\' + name
        dest = VALID_DATA_CLASS_TO + r'\\' + name
        print(src,dest)
        shutil.copy(src,dest)
                
    ###################
    # Class : From
    ###################
    allFileNames = os.listdir(HAND_PICKED_FROM)
    np.random.shuffle(allFileNames)
    trainFileNames, validFileNames = np.split(np.array(allFileNames),[int(len(allFileNames)*0.5)])
  
    print('Total hand picked from: ', len(allFileNames))
    print('Training: ', len(trainFileNames))
    print('Validation: ', len(validFileNames))

    # Copy-images to correct folder (train, valid) for this class
    
    # TRAIN FROM
    for name in trainFileNames:
        src = HAND_PICKED_FROM + r'\\' + name
        dest = TRAIN_DATA_CLASS_FROM + r'\\' + name
        print(src,dest)
        shutil.copy(src,dest)

    # VALID FROM    
    for name in validFileNames:
        src = HAND_PICKED_FROM + r'\\' + name
        dest = VALID_DATA_CLASS_FROM + r'\\' + name
        print(src,dest)
        shutil.copy(src,dest)
        
#def zipData():
    
   #shutil.make_archive(DATA_ROOT, 'zip', DATA_ROOT, 'mike_from_to')
```

```python
# create folders
createFolder(TEST_DATA)
createFolder(TRAIN_DATA_CLASS_TO)
createFolder(TRAIN_DATA_CLASS_FROM)
createFolder(VALID_DATA_CLASS_TO)
createFolder(VALID_DATA_CLASS_FROM)
print("Folders created")
```

```python
#Create Test Data
createTestData(ALL_TRACTORS, HAND_PICKED,TEST_DATA)
print("Test data created")
```

```python
#Create Train and Validate Data
createTrainAndValidateData(TRAIN_DATA_CLASS_TO, TRAIN_DATA_CLASS_FROM, VALID_DATA_CLASS_TO, VALID_DATA_CLASS_FROM)
print("Training and Validation data created")
```
