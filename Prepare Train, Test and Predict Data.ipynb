```python
# create a test data folders which
# have both from and to tractors
# but do not have any of the images selected for train and validate
 
train_images = 200    
split_train_test = 0.8    

# Source data - manually classified
CLASSIFIED = r'C:\DS Project\_processed\Images\resized_255_sq\tractors'
MIKE_CLASSIFIED = r'C:\DS Project\_processed\Images\resized_255_sq\tractors\mike'
NEIL_CLASSIFIED = r'C:\DS Project\_processed\Images\resized_255_sq\tractors\neil'

MIKE_CLASSIFIED_TO = CLASSIFIED + r'\mike\to'
MIKE_CLASSIFIED_FROM = CLASSIFIED + r'\mike\from'
NEIL_CLASSIFIED_TO = CLASSIFIED + r'\neil\to'
NEIL_CLASSIFIED_FROM = CLASSIFIED + r'\neil\from'

##################################
# folder structure for CNN input
##################################
DATA_ROOT = r'C:\DS Project\_for_keras'

## MIKE 
MIKE_ROOT = DATA_ROOT + r'\mike\data'

# Train
TRAIN_DATA_MIKE = DATA_ROOT + r'\mike\data\train'
TRAIN_DATA_MIKE_TO = TRAIN_DATA_MIKE + r'\to'
TRAIN_DATA_MIKE_FROM = TRAIN_DATA_MIKE + r'\from'

# Test
TEST_DATA_MIKE = DATA_ROOT + r'\mike\data\test' #for validation
TEST_DATA_MIKE_TO = TEST_DATA_MIKE + r'\to'
TEST_DATA_MIKE_FROM = TEST_DATA_MIKE + r'\from'

# Predict
MIKE_PREDICT = DATA_ROOT + r'\mike\data\tractor\predict'

## NEIL
NEIL_ROOT = DATA_ROOT + r'\neil\data'

# Train
TRAIN_DATA_NEIL = DATA_ROOT + r'\neil\data\train'
TRAIN_DATA_NEIL_TO = TRAIN_DATA_NEIL + r'\to'
TRAIN_DATA_NEIL_FROM = TRAIN_DATA_NEIL + r'\from'

# Test
TEST_DATA_NEIL = DATA_ROOT + r'\neil\data\test' #for validation
TEST_DATA_NEIL_TO = TEST_DATA_NEIL + r'\to'
TEST_DATA_NEIL_FROM = TEST_DATA_NEIL + r'\from'

# Predict
NEIL_PREDICT = DATA_ROOT + r'\neil\data\tractor\predict'
```

```python
import os
import shutil 
import numpy as np
```

```python
def createFolder(folder):
    
    try:
        os.makedirs(folder)
    except OSError:
        print ("Creation of the directory %s failed" % folder)
    else:
        print ("Successfully created the directory %s " % folder)
   
          
def createPredictData(resized, prepared, predict):
       
    #get all files in the test and train folders
    preppedfiles = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(prepared):
        for file in f:
            preppedfiles.append(file)

    print (len(preppedfiles))            

    #get all files in the original resized folders
    #resizedfiles = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(resized):
        for resizedfilename in f:
            
            if resizedfilename not in preppedfiles:
                #copy file to predict folder
                src = r + os.sep + resizedfilename
                dest = predict + r'\\' + resizedfilename
                print (src, dest)
                shutil.copy(src, dest)            
    
            
def createTrainAndTestData(class_to, class_from, train_to, test_to, train_from, test_from):
    
    #split the To Digester class images into train and validate
    
    ###################
    # Class : To
    ###################
    allFileNames = os.listdir(class_to)
    np.random.shuffle(allFileNames)
    
    #pick random subset of images
    total_to_select = int((1/split_train_test) * train_images)
    print (total_to_select)
    random_files = np.random.choice(allFileNames,total_to_select,replace=False)

    #split subset into train and test
    trainFileNames, testFileNames = np.split(np.array(random_files),[int(len(random_files)*split_train_test)])
  
    print('All: ', len(allFileNames))
    print('Total TO selected: ', len(random_files))
    print('Train: ', len(trainFileNames))
    print('Test: ', len(testFileNames))

    # Copy-images to correct folder (train, test) for this class
    # TRAIN TO
    for name in trainFileNames:
        src = class_to + r'\\' + name
        dest = train_to + r'\\' + name
        #print("train",src,dest
        #print(src)      
        shutil.copy(src,dest)

    print ("------")    
        
    # TEST TO    
    for name in testFileNames:
        src = class_to + r'\\' + name
        dest = test_to + r'\\' + name
        #print("test",src,dest)
        #print(name)
        shutil.copy(src,dest)
                
    ###################
    # Class : From
    ###################
    allFileNames = os.listdir(class_from)
    np.random.shuffle(allFileNames)
    
    total_to_select = int((1/split_train_test) * train_images)
    random_files = np.random.choice(allFileNames,total_to_select,replace=False)

    #split subset into train and test
    trainFileNames, testFileNames = np.split(np.array(random_files),[int(len(random_files)*split_train_test)])
  
    print('Total FROM selected: ', len(random_files))
    print('Train: ', len(trainFileNames))
    print('Test: ', len(testFileNames))

    # Copy-images to correct folder (train, valid) for this class
    
    # TRAIN FROM
    for name in trainFileNames:
        src = class_from + r'\\' + name
        dest = train_from + r'\\' + name
        #print("train",src,dest)
        shutil.copy(src,dest)

    # TEST FROM    
    for name in testFileNames:
        src = class_from + r'\\' + name
        dest = test_from + r'\\' + name
        #print("test",src,dest)
        shutil.copy(src,dest)
        
#def zipData():
    
   #shutil.make_archive(DATA_ROOT, 'zip', DATA_ROOT, 'mike_from_to')
```

```python
# create folders
createFolder(MIKE_TO_PREDICT)
createFolder(TRAIN_DATA_MIKE_TO)
createFolder(TRAIN_DATA_MIKE_FROM)
createFolder(TEST_DATA_MIKE_TO)
createFolder(TEST_DATA_MIKE_FROM)

createFolder(NEIL_TO_PREDICT)
createFolder(TRAIN_DATA_NEIL_TO)
createFolder(TRAIN_DATA_NEIL_FROM)
createFolder(TEST_DATA_NEIL_TO)
createFolder(TEST_DATA_NEIL_FROM)
print("Folders created")
```

```python
#create train data (specified number of random images To and From)
#and validate data (specific no random images, not selected for train)

#mike
createTrainAndTestData(MIKE_CLASSIFIED_TO,MIKE_CLASSIFIED_FROM,TRAIN_DATA_MIKE_TO,TEST_DATA_MIKE_TO,TRAIN_DATA_MIKE_FROM,TEST_DATA_MIKE_FROM)
createPredictData(MIKE_CLASSIFIED, MIKE_ROOT, MIKE_PREDICT)

#neil
createTrainAndTestData(NEIL_CLASSIFIED_TO,NEIL_CLASSIFIED_FROM,TRAIN_DATA_NEIL_TO,TEST_DATA_NEIL_TO,TRAIN_DATA_NEIL_FROM,TEST_DATA_NEIL_FROM)
createPredictData(NEIL_CLASSIFIED, NEIL_ROOT, NEIL_PREDICT)

print("Train, test, predict data created")
```
